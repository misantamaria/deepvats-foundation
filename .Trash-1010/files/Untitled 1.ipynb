```python
import pandas as pd

# Crear DataFrame con los factores y listas
data = {
    'Factor1': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Factor2': [['X'], ['X'], ['X', 'Y'], ['X', 'Y'], ['X', 'Y', 'Z'], ['X', 'Y', 'Z']]
}

df = pd.DataFrame(data)


```

```python
df
```

```python
# Explorar la columna 'Factor2' para crear filas separadas para cada elemento de la lista
df_exploded = df.explode('Factor2')
df_exploded
```

```python
# Crear una tabla de contingencia
crosstab_result = pd.crosstab(df_exploded['Factor1'], df_exploded['Factor2'])
crosstab_result
```

```python
# Normalizar las filas para que sumen 100
crosstab_normalized = crosstab_result.div(crosstab_result.sum(axis=1), axis=0) * 100

# Mostrar el resultado
print(crosstab_normalized)
```
